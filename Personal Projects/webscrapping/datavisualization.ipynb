```
def run_query(query):
    db = connect_to_database()
    cursor = db.cursor()
    cursor.execute(query)
    result = cursor.fetchall()
    cursor.close()
    db.close()
    return pd.DataFrame(result)
```

```
query1 = """
SELECT `country`, `CO2 Emissions(tons)` FROM `CO2`
LIMIT 10;
"""
df1 = run_query(query1)
plt.figure(figsize=(10, 8))
plt.barh(df1[0], df1[1], color='skyblue') 
plt.xlabel('CO2 Emissions')
plt.ylabel('Country')
plt.title('Los 10 países con mas emisiones')
plt.gca().invert_yaxis()
plt.show()
```

```
query2 = """
SELECT `country`,`population(2016)`, `CO2 Emissions(tons)` FROM `CO2`
LIMIT 15;
"""
df2 = run_query(query2)
plt.figure(figsize=(10, 8))
sns.scatterplot(x=df2[1], y=df2[2],hue=df2[0])
plt.xlabel('population ')
plt.ylabel('Co2 Emissions')
plt.title('Relacion de poblacion con nivel de CO2')
plt.show()
```
