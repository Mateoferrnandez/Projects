```python
def run_query(query):
    # This function runs a given SQL query on the connected database and returns the result as a pandas DataFrame
    
    db = connect_to_database()  # Establish connection to the database
    cursor = db.cursor()  # Create a cursor object to interact with the database
    
    cursor.execute(query)  # Execute the provided SQL query
    result = cursor.fetchall()  # Fetch all the results of the query
    
    cursor.close()  # Close the cursor to free up resources
    db.close()  # Close the database connection
    
    return pd.DataFrame(result)  # Convert the result to a pandas DataFrame and return it
```

## Top CO2 Emitters in the World

```python
# Define the SQL query to select country names and CO2 emissions from the CO2 table, limiting the results to the first 10 rows
query1 = """
SELECT `country`, `CO2 Emissions(tons)` FROM `CO2`
LIMIT 10;
"""

df1 = run_query(query1)  # Execute the query and store the result in a DataFrame

# Plotting the data using matplotlib
plt.figure(figsize=(10, 8))  # Create a new figure with a specified size
plt.barh(df1[0], df1[1], color='skyblue')  # Create a horizontal bar chart with country names on the y-axis and CO2 emissions on the x-axis

# Adding labels and title to the plot
plt.xlabel('CO2 Emissions')  # Label for the x-axis
plt.ylabel('Country')  # Label for the y-axis
plt.title('Countries with the Most CO2 Emissions')  # Title of the plot

plt.gca().invert_yaxis()  # Invert the y-axis to have the country with the highest emissions at the top
plt.show()  # Display the plot
```

## Relationship Between Population and CO2 Emissions

```python
# Define the SQL query to select country names, population, and CO2 emissions from the CO2 table, limiting the results to the first 15 rows
query2 = """
SELECT `country`, `population(2016)`, `CO2 Emissions(tons)` FROM `CO2`
LIMIT 15;
"""

df2 = run_query(query2)  # Execute the query and store the result in a DataFrame

# Plotting the data using seaborn
plt.figure(figsize=(10, 8))  # Create a new figure with a specified size

# Create a scatter plot with population on the x-axis, CO2 emissions on the y-axis, and color-coded by country
sns.scatterplot(x=df2[1], y=df2[2], hue=df2[0])

# Adding labels and title to the plot
plt.xlabel('Population')  # Label for the x-axis
plt.ylabel('CO2 Emissions')  # Label for the y-axis
plt.title('Relation of Population with CO2 Emissions')  # Title of the plot

plt.show()  # Display the plot
```

Claramente existe una relacin directamente proporcional entre la cantidad de CO2 emitida y la poblacin .
