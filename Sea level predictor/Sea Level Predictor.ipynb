# Sea Level Predictor
You will analyze a dataset of the global average sea level change since 1880. You will use the data to predict the sea level change through year 2050.

Use the data to complete the following tasks:

1. Use Pandas to import the data from epa-sea-level.csv.
2. Use matplotlib to create a scatter plot using the Year column as the x-axis and the CSIRO Adjusted Sea Level column as the y-axis.
3. Use the linregress function from scipy.stats to get the slope and y-intercept of the line of best fit. Plot the line of best fit over the top of the scatter plot. Make the line go through the year 2050 to predict the sea level rise in 2050.
4. Plot a new line of best fit just using the data from year 2000 through the most recent year in the dataset. Make the line also go through the year 2050 to predict the sea level rise in 2050 if the rate of rise continues as it has since the year 2000.
The x label should be Year, the y label should be Sea Level (inches), and the title should be Rise in Sea Level.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import linregress
import numpy as np
```

#### 1. Use Pandas to import the data from epa-sea-level.csv.

```python
df=pd.read_csv('epa-sea-level.csv')
```

#### 2. Use matplotlib to create a scatter plot using the Year column as the x-axis and the CSIRO Adjusted Sea Level column as the y-axis.

```python
plt.scatter(x='Year',y='CSIRO Adjusted Sea Level',data=df)
```

#### 3. Use the linregress function from scipy.stats to get the slope and y-intercept of the line of best fit. Plot the line of best fit over the top of the scatter plot. 
#### Make the line go through the year 2050 to predict the sea level rise in 2050.


```python


years=pd.Series(df['Year'])
CSIRO_Adjusted_Sea_Level = pd.Series(df['CSIRO Adjusted Sea Level'])
slope, intercept,_,_,_=linregress(years,CSIRO_Adjusted_Sea_Level)
years_1=pd.Series(np.arange(1875,2051).tolist())
bestfit = slope * years_1 + intercept
year2050=2050
predicted2050 = slope * 2050 + intercept
print(predicted2050)
```

```python
sns.scatterplot(x='Year',y='CSIRO Adjusted Sea Level',data=df)
plt.plot(years_1,bestfit,color= 'r')
plt.scatter(year2050,predicted2050)
plt.xlabel('Year')
plt.ylabel('Sea Level (inches)')
plt.title("Rise in Sea Level.")
```

#### 4. Plot a new line of best fit just using the data from year 2000 through the most recent year in the dataset. Make the line also go through the year 2050 to predict the sea level rise in 2050 if the rate of rise continues as it has since the year 2000.

```python
#Plot a new line of best fit just using the data from year 2000 through the most recent year in the dataset. 
#Make the line also go through the year 2050 to predict the sea level rise in 2050 if the rate of rise continues as it has since the year 2000.


years2000 = df[df['Year'] >= 2000]
years2=pd.Series(years2000['Year'])
CSIRO_Adjusted_Sea_Level2 = pd.Series(years2000['CSIRO Adjusted Sea Level'])
slope, intercept,_,_,_=linregress(years2,CSIRO_Adjusted_Sea_Level2)
years_2 =pd.Series(np.arange(2000,2051).tolist())
bestfit2 = slope * years_2 + intercept
year2050=2050
predicted2050 = slope * 2050 + intercept
print(predicted2050)
```

```python

sns.scatterplot(x='Year',y='CSIRO Adjusted Sea Level',data=years2000)
plt.scatter(year2050,predicted2050)
plt.plot(years_2,bestfit2,color= 'r')
plt.xlabel('Year')
plt.ylabel('Sea Level (inches)')
plt.title("Rise in Sea Level.")
```
